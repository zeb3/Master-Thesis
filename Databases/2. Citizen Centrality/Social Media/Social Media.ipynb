# CREATE THE SOCIAL MEDIA INDEX

```python
# MERGE EACH SOCIAL NETWORK DF
result1 = pd.merge(df1, df2, on=['CITIES'])
result= pd.merge(result1, df3, on=['CITIES'])
df= pd.merge(result, COVID_Index, on=['CITIES'])
df
```

```python
# COMPUTE THE OVERALL NUMBER OF FOLLOWERS ON SOCIAL MEDIA FOR EACH CITY
df["Media_Followers"]=(df["YT_Subscribers"] + df["Fb_Followers"] + df["Tw_Followers"])
del df["YT_Subscribers"]
del df["Fb_Followers"] 
del df["Tw_Followers"] 
del df["FollowersxLikes"] # (Not really necessary)
df
```

```python
%store -r dfpopulation_cities
df4=dfpopulation_cities
df4=df4[["CITIES","pop_tot"]]
result= pd.merge(df, df4, on=['CITIES'])
df=result
df
```

```python
# COMPUTE THE NUMBER OF TOTAL FOLLOWERS ON SOCIAL MEDIA OVER THE TOTAL POPULATION OF THE CITY
df["Followers_in_pop"]=df["Media_Followers"]/df["pop_tot"]
del df["Media_Followers"]
del df["pop_tot"]
df
```

```python
# Apply the min max formula
df["ViewsxVideos"]=[round(((s-min(df.ViewsxVideos))/(max(df.ViewsxVideos)-min(df.ViewsxVideos)))*10,1) for s in df["ViewsxVideos"]]
df["SubxVideos"]=[round(((s-min(df.SubxVideos))/(max(df.SubxVideos)-min(df.SubxVideos)))*10,1) for s in df["SubxVideos"]]
df["LikesxFollowers"]=[round(((s-min(df.LikesxFollowers))/(max(df.LikesxFollowers)-min(df.LikesxFollowers)))*10,1) for s in df["LikesxFollowers"]]
df["FollowersxTweet"]=[round(((s-min(df.FollowersxTweet))/(max(df.FollowersxTweet)-min(df.FollowersxTweet)))*10,1) for s in df["FollowersxTweet"]]
df["COVID_Index"]=[round(((s-min(df.COVID_Index))/(max(df.COVID_Index)-min(df.COVID_Index)))*10,1) for s in df["COVID_Index"]]
df["Followers_in_pop"]=[round(((s-min(df.Followers_in_pop))/(max(df.Followers_in_pop)-min(df.Followers_in_pop)))*10,1) for s in df["Followers_in_pop"]]
# CREATE THE INDEX AS MEAN OF THE COLUMNS
df['Social_Media_index'] = (df.mean(axis=1))
df
```

```python
Social_Media_index=df[["CITIES","Social_Media_index"]]
Social_Media_index.sort_values("Social_Media_index", axis=0, ascending=False, inplace=True)
```
