# SMART GOVERNANCE INDEX

```python
import pandas as pd
%store -r Technology_Index
%store -r Citizen_Centrality_Index
%store -r Citizen_Engagement_Index
%store -r Citizen_satisfaction_Index
```

#### WE RESTORE EACH COMPONENT AND WE COMPUTE THE FINAL INDEX AS MEAN OF THE FOUR INDEX

```python
Technology_Index.sort_values(by=['Technology Index'], inplace=True, ascending=False)
Technology_Index= Technology_Index.rename(columns={'Technology Index': 'Technology_Index'})
Technology_Index
```

```python
Citizen_Centrality_Index.sort_values(by=['Citizen_Centrality_Index'], inplace=True, ascending=False)
Citizen_Centrality_Index["Citizen_Centrality_Index"]=round(Citizen_Centrality_Index["Citizen_Centrality_Index"],2)
Citizen_Centrality_Index
```

```python
Citizen_Engagement_Index.sort_values(by=['Citizen_Engagement_Index'], inplace=True, ascending=False)
Citizen_Engagement_Index
```

```python
Citizen_satisfaction_Index.sort_values(by=['Citizen_satisfaction_Index'], inplace=True, ascending=False)
Citizen_satisfaction_Index
```

```python
SGI = pd.merge(Technology_Index, Citizen_Centrality_Index, on=['CITIES'])
SGI = pd.merge(SGI, Citizen_Engagement_Index, on=['CITIES'])
SGI = pd.merge(SGI, Citizen_satisfaction_Index, on=['CITIES'])
SGI["Smart_Governance_Index_(SGI)"]= round(SGI.mean(axis=1),2)
SGI.sort_values(by=['Smart_Governance_Index_(SGI)'], inplace=True, ascending=False)
SGI
```

```python
%store SGI
```

```python
Technology_Index.to_excel("Technology_Index.xlsx", index=False) # "index= False" to remove index in the excel file
Citizen_Centrality_Index.to_excel("Citizen_Centrality_Index.xlsx", index=False) # "index= False" to remove index in the excel file
Citizen_Engagement_Index.to_excel("Citizen_Engagement_Index.xlsx", index=False) # "index= False" to remove index in the excel file
Citizen_satisfaction_Index.to_excel("Citizen_satisfaction_Index.xlsx", index=False) # "index= False" to remove index in the excel file
SGI.to_excel("SGI.xlsx", index=False) # "index= False" to remove index in the excel file
```
