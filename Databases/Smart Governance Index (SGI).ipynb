# SMART GOVERNANCE INDEX

```python
import pandas as pd
import folium #It allows us to make maps
```

```python
%store -r Technology_Index
%store -r Citizen_Centrality_Index
%store -r Citizen_Engagement_Index
%store -r Citizen_satisfaction_Index
```

#### WE RESTORE EACH COMPONENT AND COMPUTE THE FINAL INDEX AS MEAN OF THE FOUR INDEX

```python
Technology_Index.sort_values(by=['Technology Index'], inplace=True, ascending=False)
Technology_Index= Technology_Index.rename(columns={'Technology Index': 'Technology_Index'})
Technology_Index
```

```python
Citizen_Centrality_Index.sort_values(by=['Citizen_Centrality_Index'], inplace=True, ascending=False)
Citizen_Centrality_Index["Citizen_Centrality_Index"]=round(Citizen_Centrality_Index["Citizen_Centrality_Index"],2)
Citizen_Centrality_Index
```

```python
Citizen_Engagement_Index.sort_values(by=['Citizen_Engagement_Index'], inplace=True, ascending=False)
Citizen_Engagement_Index
```

```python
Citizen_satisfaction_Index.sort_values(by=['Citizen_satisfaction_Index'], inplace=True, ascending=False)
Citizen_satisfaction_Index
```

```python
SGI = pd.merge(Technology_Index, Citizen_Centrality_Index, on=['CITIES'])
SGI = pd.merge(SGI, Citizen_Engagement_Index, on=['CITIES'])
SGI = pd.merge(SGI, Citizen_satisfaction_Index, on=['CITIES'])
SGI["Smart_Governance_Index_(SGI)"]= round(SGI.mean(axis=1),2)
SGI.sort_values(by=['Smart_Governance_Index_(SGI)'], inplace=True, ascending=False)
SGI
```

```python
%store SGI
```

```python
Technology_Index.to_excel("Technology_Index.xlsx", index=False) # "index= False" to remove index in the excel file
Citizen_Centrality_Index.to_excel("Citizen_Centrality_Index.xlsx", index=False) # "index= False" to remove index in the excel file
Citizen_Engagement_Index.to_excel("Citizen_Engagement_Index.xlsx", index=False) # "index= False" to remove index in the excel file
Citizen_satisfaction_Index.to_excel("Citizen_satisfaction_Index.xlsx", index=False) # "index= False" to remove index in the excel file
SGI.to_excel("SGI.xlsx", index=False) # "index= False" to remove index in the excel file
```

# Create an interactive Map

```python
%store -r SGI
%store -r CapitalsforMap
```

```python
CapitalsforMap= CapitalsforMap.rename(columns={'City': 'CITIES'})
CapitalsforMap
```

```python
SGI.sort_values(by=['CITIES'], inplace=True, ascending=True)
result = pd.merge(SGI, CapitalsforMap, on=['CITIES'])
del result["Country"]
del result["Address"]
```

```python
south=result.loc[(result["CITIES"] == "Athens") | (result["CITIES"] == "Madrid") | (result["CITIES"] == "Rome")]
east=result.loc[(result["CITIES"] == "Prague") | (result["CITIES"] == "Warsaw") | (result["CITIES"] == "Zagreb")]
middle=result.loc[(result["CITIES"] == "Paris") | (result["CITIES"] == "Vienna") | (result["CITIES"] == "Bruxelles")]
north=result.loc[(result["CITIES"] == "Copenhagen") | (result["CITIES"] == "Helsinki") | (result["CITIES"] == "Stockholm")]
```

```python
m = folium.Map(location=[50, 10], zoom_start=3.5)
for i in range(0,len(east)):
    folium.CircleMarker(
        [east.iloc[i]['Lat'], east.iloc[i]['Lon']],  popup=(
            east.iloc[i]['CITIES'] +" SGI: "+str(east.iloc[i]['Smart_Governance_Index_(SGI)'])),
        radius=7,color="red",
        fill=True,
        fill_color="red"
    ).add_to(m)
    
for i in range(0,len(north)):
    folium.CircleMarker(
        [north.iloc[i]['Lat'], north.iloc[i]['Lon']],  popup=(
            north.iloc[i]['CITIES'] +" SGI: "+str(north.iloc[i]['Smart_Governance_Index_(SGI)'])),
        radius=7,color="blue",
        fill=True,
        fill_color="blue"
    ).add_to(m)
    
for i in range(0,len(middle)):
    folium.CircleMarker(
        [middle.iloc[i]['Lat'], middle.iloc[i]['Lon']],  popup=(
            middle.iloc[i]['CITIES'] +" SGI: "+str(middle.iloc[i]['Smart_Governance_Index_(SGI)'])),
        radius=7,color="green",
        fill=True,
        fill_color="green"
    ).add_to(m)
    
    
for i in range(0,len(south)):
    folium.CircleMarker(
        [south.iloc[i]['Lat'], south.iloc[i]['Lon']], popup=(
            south.iloc[i]['CITIES'] +" SGI: "+str(south.iloc[i]['Smart_Governance_Index_(SGI)'])),
        radius=7,color="orange",
        fill=True,
        fill_color="orange"
    ).add_to(m)
m
```
